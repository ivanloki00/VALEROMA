```python
# Carica il dataset (sostituisci con il percorso del tuo file)
df = pd.read_csv('Data_Netflix_cumsum.csv')

# Assicurati che la colonna 'Date' sia in formato datetime
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['date_added'] = df['date_added'].dt.to_period('M')
df.reset_index(drop=True, inplace=True)
df
```

```python
df_movies_cumsum = df.iloc[:,:].groupby('date_added').max()
df=df_movies_cumsum.reset_index()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd

import numpy as np 
import matplotlib.animation as animation 
from IPython import display 

# Sample data (replace this with your own data)

# Function to update the bar chart
def update(frame):
    plt.clf()
    
    # Get data for current frame
    current_data = df.iloc[frame]
    categories = current_data.index[1:]  # Exclude the 'Date' column
    
    # Plot bar chart
    plt.barh(categories, current_data[1:])
    
    plt.title(f"Bar Chart Race - {current_data['date_added'].strftime('%Y-%m-%d')}")
    plt.xlabel('Value')
    plt.ylabel('Category')

# Create the animation
fig = plt.figure(figsize=(10, 6))
ani = FuncAnimation(fig, update, frames=len(df), repeat=False)
# ani.save('animation.gif', writer='imagemagick')
plt.tight_layout()
plt.show()
ani
# Display the animation
# plt.close()  # Prevent double display of the animation
# HTML(ani.to_jshtml())
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd

import numpy as np 
import matplotlib.animation as animation 
from IPython import display 

# Sample data (replace this with your own data)
# Function to update the bar chart
def update(frame):
    plt.clf()
    
    # Get data for current frame
    current_data = df.iloc[frame]
    categories = current_data.index[1:]  # Exclude the 'Date' column
    
    # Plot bar chart
    plt.barh(categories, current_data[1:])
    
    plt.title(f"Bar Chart Race - {current_data['date_added'].strftime('%Y-%m-%d')}")
    plt.xlabel('Value')
    plt.ylabel('Category')

# Create the animation
fig = plt.figure(figsize=(10, 6))
ani = FuncAnimation(fig, update, frames=len(df), repeat=False)

# converting to an html5 video 
video = ani.to_html5_video() 
  
# embedding for the video 
html = display.HTML(video) 
  
# draw the animation 
display.display(html) 
plt.close() 
```
