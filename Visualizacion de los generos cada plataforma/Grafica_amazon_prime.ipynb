##  VISUALIZAR LOS GENEROS PARA AMAZON PRIME

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('amazon_prime_titles.csv')
df.sample(10)
```

Dividimos entre Movies and TV shows

```python
df['type'].unique()
```

```python
df_movies = df[df['type']=='Movie']
df_movies.reset_index(drop=True, inplace=True)
df_movies.info()
```

```python
type(df_movies['listed_in'][0])
```

Limpieza de los generos

```python
df_movies.loc[:,'listed_in'] = df_movies['listed_in'].apply(lambda x: x.replace('Entertainment, and Culture','Entertainment and Culture'))

df_movies.loc[:,'listed_in'] = df_movies['listed_in'].apply(lambda x: x.split(', '))
type(df_movies['listed_in'][0])
```

```python
df_ = pd.DataFrame(df_movies['listed_in'])
for i in df_['listed_in']:
    if 'and Culture' in i:
        print(i)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

dummies_movies = pd.DataFrame(mlb.fit_transform(df_movies['listed_in']), columns=mlb.classes_)

dummies_movies
```
