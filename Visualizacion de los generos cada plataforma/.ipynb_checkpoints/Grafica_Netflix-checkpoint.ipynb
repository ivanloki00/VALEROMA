```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly as px
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

Dividimos entre Movies and TV shows

```python
df['type'].unique()
```

```python
df_movies = df[df['type']=='Movie']
df_movies.reset_index(drop=True, inplace=True)
df_movies.info()
```

Limpieza de los generos, creacion de dummy columns

```python
# df_movies.loc[:,'listed_in'] = df_movies['listed_in'].apply(lambda x: x.replace(' &',','))

df_movies.loc[:,'listed_in'] = df_movies['listed_in'].apply(lambda x: x.split(', '))
type(df_movies['listed_in'][0])
```

```python
df_movies['listed_in']
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

dummies_genres = pd.DataFrame(mlb.fit_transform(df_movies['listed_in']), columns=mlb.classes_)

print('longitud:',len(dummies_genres.columns),'\ncolumnas:',dummies_genres.columns)
```

```python
df_movies = pd.concat([df_movies, dummies_genres], axis=1)
df_movies
```

Cambiar el formato de las fechas con pandas datetime method

```python
df_movies['date_added']
```

```python
series_fechas = pd.to_datetime(df_movies['date_added'].apply(lambda x: x.replace(',','')), format='%B %d %Y')
```

```python
series_fechas.min()
```

```python
series_fechas.sort_values(ascending=True)
```

```python
df_dates = pd.DataFrame()
df_dates['Dates'] = series_fechas.sort_values(ascending=True).unique()
```

```python
df_dates.reset_index(drop=True, inplace=True)
df_dates
```

Ahora tenemos el formato datetime que puede ser utilizado para presentar las graficas a lo largo del tiempo

```python
df_movies['date_added'] = series_fechas
df_movies.head()
```

```python
df_movies_ = df_movies.copy()
df_movies_.drop(columns=['type','director', 'cast', 'country', 'rating', 'duration', 'listed_in', 'description','release_year'], axis=1, inplace=True)
df_movies_.sort_values(by='date_added',inplace=True)
df_movies_.reset_index(drop=True, inplace=True)
df_movies_
```

```python
df_movies_.columns
```

```python
df_movies_cumsum = df_movies_.iloc[:,2:].groupby('date_added').sum().cumsum()
# df_movies_cumsum.reset_index(inplace=True)
df_movies_cumsum
```

Prova di utilizzo di circlify

```python
import circlify
```

```python
df_prova = pd.DataFrame(df_movies_cumsum.iloc[-15,:].reset_index())
df_prova.columns=['Genre', 'Value']
df_prova.head()
```

```python
# Compute circle positions
circles = circlify.circlify(
    df_prova['Value'].tolist(),
    show_enclosure=False,
    target_enclosure = circlify.Circle(x=0,y=0,r=1)
)

# Reverse the order of the circles to match the order of data
circles = circles[::-1]

print(circles)
```

```python
# Create a figure and one subplot
fir,ax = plt.subplots(figsize=(13,13))

# Remove axis boundaries
ax.axis('off')

# Find axis fboundaries
lim = max(max(abs(circle.x),abs(circle.y)) for circle in circles)

plt.xlim(-lim,lim)
plt.ylim(-lim,lim)
labels = df_prova['Genre'] 
# Print the circles
for circle,label in zip(circles,labels):
    x,y,r = circle
    ax.add_patch(plt.Circle((x,y),r,alpha=0.2, linewidth=2, fill=False))
    plt.annotate(label,(x,y),va='center',ha='center')
```
